## 死链检测代码

```python
query_result = session.query(WSTest).filter(WSTest.id < 11)
colu = ['链接','状态码','父级链接','采集时间']
for result in query_result:
    pd.DataFrame(columns=colu).to_excel(f'{result.name}_0.xlsx',index=False)
    
```

```python
# 分组代码
sitegroup = errorl.groupby('site_id')
# bsm1.get_group(4108230001).head()
for icp,temp in sitegroup:
    result = session.query(WSTest).filter(WSTest.icp==icp).first()
    print(result.name)
    print(temp.count())
    temp1=temp.loc[:,['url','error_code','parents_url','create_time']]
    temp1.rename(columns={'url':'链接','error_code':'状态码','parents_url':'父级链接','create_time':'采集时间'},inplace=1)
    temp1.to_excel(f'{result.name}_{len(temp)}.xlsx',index=False)
```
